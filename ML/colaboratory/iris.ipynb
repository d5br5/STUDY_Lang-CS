```
import tensorflow as tf
import pandas as pd
```

```
file = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(file)
```

```
iris = pd.get_dummies(iris)
```

```
iris.columns
```

```
cs = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
rs = iris[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
```

```
rs.shape
```

```
X = tf.keras.layers.Input(shape=[4])

H = tf.keras.layers.Dense(5, activation='swish')(X)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
model.fit(cs, rs, epochs=500) 
```
