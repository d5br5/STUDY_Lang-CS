```
import tensorflow as tf
import pandas as pd
```

```
lemon_dir = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemon = pd.read_csv(lemon_dir)
lemon.head()
```

```
cs = lemon[['온도']]
rs = lemon[['판매량']]
print(cs.shape, rs.shape)
```

```
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
model.fit(cs, rs, epochs=5)
```

```
model.predict(cs)
```

```
model.predict([[15]])
```
